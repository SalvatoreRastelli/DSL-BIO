```python
import os
import re
import pandas as pd

# Definizione del percorso principale del dataset e delle cartelle dei giorni di infestazione
dataset_path = '/content/Processed_Im'
day_folders = ['Day4', 'Day5', 'Day7', 'Day10']

# Pattern per estrarre le informazioni dal nome delle sottocartelle
subfolder_pattern = re.compile(r'CRI_D?(\d+)_(\d+|(?:cntrl|ctrl|contr|Ctrl)).+')

# Inizializzazione delle liste per i dati estratti
subfolder_paths = []
labels = []
days_of_infestation = []
number_of_flies = []

# Iterazione attraverso le cartelle dei giorni di infestazione
for day_folder in day_folders:
    folder_path = os.path.join(dataset_path, day_folder)

    # Iterazione attraverso le sottocartelle nella cartella del giorno
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)

        # Controllo per saltare i file nascosti e non cartelle
        if subfolder.startswith('.') or not os.path.isdir(subfolder_path):
            continue

        # Controllo e classificazione della sottocartella
        match = subfolder_pattern.match(subfolder)
        if match:
            day_of_infestation = int(match.group(1))
            flies_or_control = match.group(2)

            if flies_or_control in ['cntrl', 'ctrl', 'contr', 'Ctrl']:
                label = 'not_infested'
                flies = 0
                day_of_infestation = 0
            else:
                label = 'infested'
                flies = int(flies_or_control)
        else:
            continue

        subfolder_paths.append(subfolder_path)
        labels.append(label)
        days_of_infestation.append(day_of_infestation)
        number_of_flies.append(flies)

# Creazione del DataFrame
df = pd.DataFrame({
    'Subfolder': subfolder_paths,
    'Days of infestation': days_of_infestation,
    'Number of flies': number_of_flies,
    'Label': labels
})

# Visualizza il DataFrame
df
```

```python
df['flip'] = 0
df['rotate'] = 0
df_flip = df.copy()
df_flip['flip'] = 1
df=pd.concat([df, df_flip])
df_rotate=df.copy()
df_rotate['rotate'] = 1
df=pd.concat([df, df_rotate])
df.reset_index(drop=True, inplace=True)
df
```

```python
from PIL import Image, ImageOps
import glob

img_size = 96

for i in range(len(df)):
  # Load the images
  image_folder = df['Subfolder'][i]
  flip = df['flip'][i]
  rotate = df['rotate'][i]
  days = df['Days of infestation'][i]
  number = df['Number of flies'][i]
  label = df['Label'][i]

  # Create a list of Image objects
  images = []
  for j in range(0, 360, 5):
    filename = f'{image_folder}/vertical_section_{str(j)}deg.jpg'
    frame = Image.open(filename)
    frame = frame.resize((img_size, img_size))

    if flip == 1:
      frame = ImageOps.mirror(frame)
    elif rotate == 1:
      frame = frame.rotate(180)
    images.append(frame)

  # Save the frames as a GIF
  output_path = f'/content/gifs/lemon_{i}_{days}_{number}_{label}.gif'
  images[0].save(output_path, save_all=True, append_images=images[1:], duration=100, loop=0)

  print(f"GIF saved to {output_path}")
```
